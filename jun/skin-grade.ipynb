```python
import cv2
import matplotlib.pyplot as plt
import os
import json

# JSON 파일 경로 설정
base_dir = r"D:\data\korean skin data\open data\data\Training\Label data\digiter cam"
# 이미지 파일 경로 설정
image_base_path = r"D:\data\korean skin data\open data\data\Training\Raw data\digiter cam"
# 출력 폴더 경로 설정
output_base_dir = r"D:\data\korean 01 data"

start_id = 2  
end_id = 1100 

# data 딕셔너리에 JSON 파일 데이터 로드
data = {}
for i in range(start_id, end_id + 1):
    name_value = f'{i:04d}'
    json_file_names = f"{name_value}_01_F_01.json"

    file_path = os.path.join(base_dir, name_value, json_file_names)
    if not os.path.exists(file_path):
        continue

    with open(file_path, 'r', encoding='utf-8') as f:
        data[json_file_names] = json.load(f)

# bbox_list에 bbox 정보 저장
bbox_list = []
for key, value in data.items():
    if 'images' in value and 'bbox' in value['images']:
        bbox = value['images']['bbox']
        bbox_list.append((key, bbox))

# 각 bbox에 대해 이미지 자르고 저장
for key, bbox in bbox_list:
    id_value = key.split('_')[0]  # ID 추출
    image_path = os.path.join(image_base_path, id_value, f"{id_value}_01_F.jpg")
    
    if not os.path.exists(image_path):
        continue

    # 이미지 읽기
    image = cv2.imread(image_path)

    # 이미지 자르기
    cropped_image = image[bbox[1]:bbox[3], bbox[0]:bbox[2]]

    # 출력 폴더 생성
    output_dir = os.path.join(output_base_dir, id_value)
    os.makedirs(output_dir, exist_ok=True)

    # 자른 이미지 저장 경로
    output_path = os.path.join(output_dir, f"cropped_{key}.jpg")
    
    # 자른 이미지 저장
    cv2.imwrite(output_path, cropped_image)

    # 자른 이미지 확인 (옵션)
    plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```
